# Codage source

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('../../_static/graph.mplstyle')
```

## Auto-information

```python
p = np.logspace(-7, 0, 100)
I = - np.log2(p)
```

```python
fig, ax = plt.subplots(1,1, figsize=(3,3))
plt.ylim([0, 20])
plt.plot(p, I)
plt.xlabel("$p_m$")
plt.ylabel("$I(x_m) \;$ [Sh]")
plt.xticks(np.arange(0, 1.1, .25))
plt.yticks(np.arange(0, 20, 5))
plt.savefig("auto-information.svg")
plt.show()
```

## Entropie

```python
d = 1e-2
p = np.arange(d, 1, d)
p = np.concatenate(([1e-9], p, [1-1e-9]))
H = -p*np.log2(p) - (1-p)*np.log2(1-p)
```

```python
fig, ax = plt.subplots(1,1, figsize=(3,3))
plt.ylim([-.05, 1.05])
plt.plot(p, H)
plt.xlabel(r"$\alpha$")
plt.ylabel("$H \;$ [Sh/symb]")
plt.xticks(np.arange(0, 1.1, .25))
plt.yticks(np.arange(0, 1.1, .25))
plt.savefig("entropie.svg")
plt.show()
```
